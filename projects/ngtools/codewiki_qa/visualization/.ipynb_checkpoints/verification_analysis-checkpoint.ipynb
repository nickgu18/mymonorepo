```python
import json
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
def analyze_verification_results(directory):
    verification_files = glob.glob(os.path.join(directory, '*_verification_ans*.json'))
    results = []

    for verification_file in verification_files:
        question_id = os.path.basename(verification_file).split('_verification_ans')[0]
        answer_id = os.path.basename(verification_file).split('.json')[0].split('_verification_ans')[1]
        result_file = os.path.join(directory, f"{question_id}_verify_result_ans{answer_id}.json")

        if not os.path.exists(result_file):
            print(f"Result file not found for {verification_file}")
            continue

        with open(verification_file, 'r') as f:
            verification_data = json.load(f)

        with open(result_file, 'r') as f:
            result_data = json.load(f)

        question_text = verification_data.get('original_question', 'N/A')

        if 'verification_questions' in verification_data:
            for i, verification in enumerate(verification_data['verification_questions']):
                results.append({
                    'question_id': question_id,
                    'question': question_text,
                    'verification_question': verification,
                    'decision': result_data['verification_results'][i]['decision'],
                    'rationale': result_data['verification_results'][i]['rationale']
                })
        else:
            results.append({
                    'question_id': question_id,
                    'question': question_text,
                    'verification_question': verification_data['verification_question'],
                    'decision': result_data['decision'],
                    'rationale': result_data['rationale']
                })

    return pd.DataFrame(results)
```

```python
def visualize_pass_rate(df):
    if df.empty:
        print("No data to visualize.")
        return
    
    for question_id, group in df.groupby('question_id'):
        try:
            
            main_question = group['question'].iloc[0]
            pass_rate = group['decision'].apply(lambda x: 1 if x.lower() == 'pass' else 0).mean()
            fail_rate = 1 - pass_rate
            plt.figure(figsize=(6, 4))
            plt.bar(['Pass', 'Fail'], [pass_rate, fail_rate], color=['green', 'red'])
            plt.title(f'Verification Pass Rate for Question ID: {question_id}')
            plt.ylabel('Percentage')
            plt.ylim(0, 1)
            plt.show()
        except Exception as e:
            print ("visualization failed")
        
        print(f"Question ID: {question_id}")
        print(f"Question: {main_question}")
        print(f"Verification Details:")
        html = ''
        for _, row in group.iterrows():
            html += f"<details><summary><b>{row['verification_question']}</b></summary>"
            html += f"<p><b>Decision:</b> {row['decision']}</p>"
            html += f"<p><b>Rationale:</b> {row['rationale']}</p></details>"
        display(HTML(html))
```
