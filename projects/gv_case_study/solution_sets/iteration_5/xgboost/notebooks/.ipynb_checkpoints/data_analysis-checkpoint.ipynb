# Training Data Analysis

Exploratory analysis of training founders, company coverage, and industries.

```python
import sys
from pathlib import Path
from dataclasses import dataclass
from typing import Iterable

import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.common import load_config
from src.data import load_raw, load_targets


@dataclass(frozen=True)
class TrainingCoverageSummary:
    """Summary of how well founders' companies are covered by training metadata."""

    unique_founders: int
    founders_with_unmatched_companies: int
    unmatched_company_ids: int


@dataclass(frozen=True)
class FounderCompanyCoverageSummary:
    """Coverage summary focused on founders (is_founder == True)."""

    founders_with_founder_role: int
    founders_with_company_in_targets: int
    founders_with_company_in_metadata: int
    founders_with_all_companies_unmatched: int
    unique_founding_companies: int
    unmatched_founding_companies: int


def summarize_training_company_coverage(
    experience: pd.DataFrame,
    company_info: pd.DataFrame,
    targets: pd.DataFrame,
) -> TrainingCoverageSummary:
    """Return coverage summary for founders vs company_info/targets.

    A founder is counted as having unmatched companies if any of their
    experience rows reference a company_id that is missing from both
    company_info and target_variable_training.
    """

    exp = experience.copy()
    ci = company_info.copy()
    tgt = targets.copy()

    if "company_id" not in exp.columns:
        raise ValueError("experience is missing 'company_id'")
    if "person_id" not in exp.columns:
        raise ValueError("experience is missing 'person_id'")

    exp_company_ids = set(exp["company_id"].dropna().unique())
    ci_company_ids = set(ci.get("company_id", pd.Series([], dtype=object)).dropna().unique())
    tgt_company_ids = set(tgt.get("company_id", pd.Series([], dtype=object)).dropna().unique())

    covered_company_ids = ci_company_ids | tgt_company_ids
    unmatched_company_ids = exp_company_ids - covered_company_ids

    mask_unmatched = exp["company_id"].isin(unmatched_company_ids)
    founders_with_unmatched = exp.loc[mask_unmatched, "person_id"].dropna().nunique()
    unique_founders = exp["person_id"].dropna().nunique()

    return TrainingCoverageSummary(
        unique_founders=unique_founders,
        founders_with_unmatched_companies=founders_with_unmatched,
        unmatched_company_ids=len(unmatched_company_ids),
    )


def summarize_founder_company_coverage(
    experience: pd.DataFrame,
    company_info: pd.DataFrame,
    targets: pd.DataFrame,
) -> FounderCompanyCoverageSummary:
    """Return coverage summary restricted to founders (is_founder == True).

    A founder is included if they have at least one experience row with
    is_founder == True. A company is considered present in training data if it
    appears in company_info or target_variable_training.
    """

    if "company_id" not in experience.columns:
        raise ValueError("experience is missing 'company_id'")
    if "person_id" not in experience.columns:
        raise ValueError("experience is missing 'person_id'")
    if "is_founder" not in experience.columns:
        raise ValueError("experience is missing 'is_founder'")

    exp = experience.copy()

    flag = exp["is_founder"]
    if flag.dtype is not bool:
        flag = flag.astype(str).str.strip().str.lower().isin({"true", "1", "yes", "y", "t"})
    exp["is_founder"] = flag

    founders_exp = exp[exp["is_founder"]].dropna(subset=["person_id", "company_id"])
    founders_with_founder_role = founders_exp["person_id"].nunique()

    if founders_with_founder_role == 0:
        return FounderCompanyCoverageSummary(
            founders_with_founder_role=0,
            founders_with_company_in_targets=0,
            founders_with_company_in_metadata=0,
            founders_with_all_companies_unmatched=0,
            unique_founding_companies=0,
            unmatched_founding_companies=0,
        )

    ci_company_ids = set(
        company_info.get("company_id", pd.Series([], dtype=object)).dropna().unique()
    )
    tgt_company_ids = set(
        targets.get("company_id", pd.Series([], dtype=object)).dropna().unique()
    )
    covered_company_ids = ci_company_ids | tgt_company_ids

    targets_company_ids = tgt_company_ids

    grouped = founders_exp.groupby("person_id")["company_id"].agg(lambda col: set(col.dropna()))

    founders_with_company_in_targets = 0
    founders_with_company_in_metadata = 0
    founders_with_all_companies_unmatched = 0

    founding_companies_all: set[str] = set()
    founding_companies_unmatched: set[str] = set()

    for company_ids in grouped:
        if not company_ids:
            continue
        founding_companies_all.update(company_ids)

        has_in_targets = bool(company_ids & targets_company_ids)
        has_in_metadata = bool(company_ids & covered_company_ids)
        all_unmatched = not has_in_metadata

        if has_in_targets:
            founders_with_company_in_targets += 1
        if has_in_metadata:
            founders_with_company_in_metadata += 1
        if all_unmatched:
            founders_with_all_companies_unmatched += 1

        unmatched_here = company_ids - covered_company_ids
        founding_companies_unmatched.update(unmatched_here)

    return FounderCompanyCoverageSummary(
        founders_with_founder_role=founders_with_founder_role,
        founders_with_company_in_targets=founders_with_company_in_targets,
        founders_with_company_in_metadata=founders_with_company_in_metadata,
        founders_with_all_companies_unmatched=founders_with_all_companies_unmatched,
        unique_founding_companies=len(founding_companies_all),
        unmatched_founding_companies=len(founding_companies_unmatched),
    )


def build_founder_experience_summary(
    experience: pd.DataFrame, company_info: pd.DataFrame
) -> pd.DataFrame:
    """Return per-founder summary of company experiences and industries.

    The result has one row per founder_id with:
    - company_count: number of unique companies in their experience history
    - industries: sorted, comma-separated list of industries for those companies
    """

    if "person_id" not in experience.columns:
        raise ValueError("experience is missing 'person_id'")
    if "company_id" not in experience.columns:
        raise ValueError("experience is missing 'company_id'")
    if "company_id" not in company_info.columns:
        raise ValueError("company_info is missing 'company_id'")
    if "industry" not in company_info.columns:
        raise ValueError("company_info is missing 'industry'")

    exp = experience[["person_id", "company_id"]].dropna(subset=["person_id", "company_id"]).copy()
    ci = company_info[["company_id", "industry"]].copy()

    exp = exp.merge(ci, on="company_id", how="left")

    def _join_sorted(values: Iterable[str]) -> str:
        unique = sorted({v for v in values if isinstance(v, str) and v})
        return ", ".join(unique)

    grouped = (
        exp.groupby("person_id")
        .agg(
            company_count=("company_id", lambda col: col.dropna().nunique()),
            industries=("industry", _join_sorted),
        )
        .reset_index()
        .rename(columns={"person_id": "founder_id"})
    )

    return grouped
```

```python
cfg = load_config()
raw_training = load_raw(cfg.data, dataset='training')
targets = load_targets(cfg.data)
raw_training.experience.head()
```

## 1. Coverage of founders' companies vs training metadata

```python
coverage = summarize_training_company_coverage(
    experience=raw_training.experience,
    company_info=raw_training.company_info,
    targets=targets,
)
coverage
```

## 2. Founder multi-company experience and industries

```python
founder_summary = build_founder_experience_summary(
    experience=raw_training.experience,
    company_info=raw_training.company_info,
)
multi_company = founder_summary[founder_summary['company_count'] > 1]
len(multi_company), founder_summary.head()
```

## 3. Founder (is_founder=True) company coverage

```python
founder_cov = summarize_founder_company_coverage(
    experience=raw_training.experience,
    company_info=raw_training.company_info,
    targets=targets,
)
founder_cov
```
